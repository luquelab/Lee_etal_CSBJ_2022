```python
# function to clean up an NCBI alignment text file

def newAlign(sim_table):
    
    # rename the columns
    sim_table.columns = ['query from', 'query to', 'percent identity', 'alignment length','mismatches', 'gap opens', 'query start', 'query end', 's start', 's end', 'evalue', 'bit score', 'percent positives']
    # create a boolean for each line that is true if it starts with 'ph'
    is_ph =  sim_table['query from']>'ph'
    # create a new table based on that boolean filter
    sim_ph = sim_table[is_ph]
    # create a boolean for each line that compares to itself
    is_nxn =  sim_ph['query from']!=sim_ph['query to']
    # apply the second filter
    sim_ph = sim_ph[is_nxn]
    # strip the 'ph' from the table
    sim_ph["query from"] = sim_ph["query from"].map(lambda x: x.lstrip('ph'))
    sim_ph["query to"] = sim_ph["query to"].map(lambda x: x.lstrip('ph'))
    sim_ph = sim_ph.sort_values(by=["query from","query to"])
    sim_ph['from_to'] = sim_ph['query from'].map(str) + '_' + sim_ph['query to'].map(str)
    sim_ph['to_from'] = sim_ph['query to'].map(str) + '_' + sim_ph['query from'].map(str)
    sim_ph = sim_ph.reset_index(drop=True)

    #print(sim_ph.shape)
    return sim_ph
    #print(sim_ph.head())
```

```python
# calculate T numbers
ny = phageData.shape[0]
Y_T = []

for i in range(ny):
    # ID
    Y_T.append(phageData.iloc[i]["Virus_ID"])
    # Check and see if the structure is on the High-Res list
    if(phageData.iloc[i]['HR data']=="x"):
        # T raw
        Y_T.append(phageData.iloc[i]['HR_T'])
        # T nearest
        Y_T.append(phageData.iloc[i]['HR_T'])
        # T nearest err mar
        Y_T.append(phageData.iloc[i]['HR_T'])
        # T nearest err mar code
        Y_T.append(tdict2[phageData.iloc[i]['HR_T']])
    else:
        # T raw
        Y_T.append(round(tNum(phageData.iloc[i]["genome_length"]/1000,0),4))
        # T nearest
        Y_T.append(tNum(phageData.iloc[i]["genome_length"]/1000,1))
        # T nearest err mar
        Y_T.append(tNum(phageData.iloc[i]["genome_length"]/1000,2,errMar))
        # T nearest err mar code
        Y_T.append(tdict2[tNum(phageData.iloc[i]["genome_length"]/1000,2,errMar)])
    
Y = np.asarray(Y_T)
Y = np.reshape(np.ravel(Y), (ny, 5));
Y = np.asarray(Y)

df_T = pd.DataFrame(Y)
df_T = df_T.rename(columns={0: 'Virus_ID', 1: 'T_raw', 2: 'T_nearest', 3: 'T_nearest_errMar', 4: 'T_nearest_errMar_code'})

df_T["T_raw"] = df_T["T_raw"].astype('float64')
df_T["T_nearest"] = df_T["T_nearest"].astype('float64')
df_T["T_nearest_errMar"] = df_T["T_nearest_errMar"].astype('float64')
df_T["T_nearest_errMar_code"] = df_T["T_nearest_errMar_code"].astype('int64')

df_T [32:40]
```

```python
# create the table of average relative difference in T-number
Tij_dist = []
for i in range(phageData.shape[0]):
    for j in range(phageData.shape[0]):
        if (j>i):
            Tij_dist.append(phageData.iloc[i]['Virus_ID'])
            Tij_dist.append(phageData.iloc[j]["Virus_ID"])
            Tij_dist.append(0.5*((abs(phageData.iloc[i]['T_raw']-phageData.iloc[j]['T_raw'])/phageData.iloc[i]['T_raw'])+(abs(phageData.iloc[i]['T_raw']-phageData.iloc[j]['T_raw'])/phageData.iloc[j]['T_raw'])))

Tij_dist2 = pd.DataFrame(np.reshape(np.ravel(Tij_dist), (-1, 3)))
Tij_dist2.columns = ['query from', 'query to','T_diff']
Tij_dist2['query from']=Tij_dist2['query from'].astype(int)
Tij_dist2['query to']=Tij_dist2['query to'].astype(int)     
#Tij_dist2.to_csv('Tij_dist2.csv', index=False)
```

```python
# set train and test sizes
n80 = round(n*.80)
n20 = n-n80

# create a random shuffle of the phage data
shuffle_index = np.random.permutation(n)

# create a random shuffle of the phage data for this iteration
shuffle_index = np.random.permutation(shuffle_index)

# split into an 80/20 training set
p80 = np.asarray(phageData["Virus_ID"][shuffle_index][:n80])
p20 = np.asarray(phageData["Virus_ID"][shuffle_index][n80:])

# break the test set off of the data frame to make the training set
countcount = []
for i in range(all_matches.shape[0]):
    if(all_matches.iloc[i]["query to"] in p20):
        #print(T_sim.iloc[i]["query to"])
        countcount.append(i)
current_train_set = copy.copy(all_matches)
for i in range (all_matches.shape[0]):
    if i in countcount:
        current_train_set = current_train_set.drop(i)
current_train_set = current_train_set.reset_index(drop=True)
```

```python
# run the test set against the training set for a T-number match
match_finder = []
    
# for each phage in the test set
for j in p20:
    # look at all matches in the training set
    current_frame = current_train_set[current_train_set['query from']==j]

    # if there are no matches, append a nonsense value (999)
    if (current_frame.shape[0]==0):
            match_finder.append(j)
            match_finder.append(999)
    # if there are matches, determine the T-number
    else:
        # find the maximum percent identity match from the training set
        max_current_matches = current_frame[current_frame["percent identity"]==max(current_frame["percent identity"])]
        #print(max_current_matches)
        # if there is only one structure in the training set with the max percent match, then use that T-number 
        if (max_current_matches.shape[0]==1):
            match_finder.append(j)
            target_T = max_current_matches.iloc[0]['query to']
            match_finder.append(float(phageData[phageData["Virus_ID"]==target_T]["T_nearest_errMar"]))
        # if there are several matches with the same max percent match, use the consensus of all of the matches for the T-number
        else:
            max_finder_match = []
            for k in range(max_current_matches.shape[0]):
                target_T = max_current_matches.iloc[k]['query to']
                max_finder_match.append(float(phageData[phageData["Virus_ID"]==target_T]["T_nearest_errMar"]))
            match_finder.append(j)
            match_finder.append(statistics.mode(max_finder_match))

p20_test_matches = pd.DataFrame(np.reshape(np.ravel(match_finder), (-1, 2)))
p20_test_matches.columns = ['Virus_ID','T_pred']

# add G2T T
p20_test_result = p20_test_matches.merge(phageData[['Virus_ID','T_nearest']], how='left', on='Virus_ID')
# check prediction
p20_test_result["Correct"]=(p20_test_result["T_nearest"]-p20_test_result["T_pred"])
# calculate accuracy
sim_accuracy = p20_test_result[p20_test_result['Correct']==0].shape[0]
print('accuracy = %.3f (%d)' % (sim_accuracy/n20, sim_accuracy))
no_prediction = p20_test_matches[p20_test_matches['T_pred']==999].shape[0]
print('no prediction = %.3f (%d)' % (no_prediction/n20, no_prediction))
```

```python
# run it ten times with no similarity restriction

# set train and test sizes
n80 = round(n*.80)
n20 = n-n80

# set number of iterations at each percent similarity
n_iter = 10
all_missing = []
all_correct = []

# create a random shuffle of the phage data
shuffle_index = np.random.permutation(n)

# loop 
for q in range(n_iter):

    # set train and test sizes
    n80 = round(n*.80)
    n20 = n-n80

    # create a random shuffle of the phage data for this iteration
    shuffle_index = np.random.permutation(shuffle_index)

    # split into an 80/20 training set
    p80 = np.asarray(phageData["Virus_ID"][shuffle_index][:n80])
    p20 = np.asarray(phageData["Virus_ID"][shuffle_index][n80:])

    # break the test set off of the data frame to make the training set
    countcount = []
    for i in range(all_matches.shape[0]):
        if(all_matches.iloc[i]["query to"] in p20):
            #print(T_sim.iloc[i]["query to"])
            countcount.append(i)
    current_train_set = copy.copy(all_matches)
    for i in range (all_matches.shape[0]):
        if i in countcount:
            current_train_set = current_train_set.drop(i)
    current_train_set = current_train_set.reset_index(drop=True)

    # run the test set against the training set for a T-number match
    match_finder = []

    # for each phage in the test set
    for j in p20:
        # look at all matches in the training set
        current_frame = current_train_set[current_train_set['query from']==j]

        # if there are no matches, append a nonsense value (999)
        if (current_frame.shape[0]==0):
                match_finder.append(j)
                match_finder.append(999)
        # if there are matches, determine the T-number
        else:
            # find the maximum percent identity match from the training set
            max_current_matches = current_frame[current_frame["percent identity"]==max(current_frame["percent identity"])]
            #print(max_current_matches)
            # if there is only one structure in the training set with the max percent match, then use that T-number 
            if (max_current_matches.shape[0]==1):
                match_finder.append(j)
                target_T = max_current_matches.iloc[0]['query to']
                match_finder.append(float(phageData[phageData["Virus_ID"]==target_T]["T_nearest_errMar"]))
            # if there are several matches with the same max percent match, use the consensus of all of the matches for the T-number
            else:
                max_finder_match = []
                for k in range(max_current_matches.shape[0]):
                    target_T = max_current_matches.iloc[k]['query to']
                    max_finder_match.append(float(phageData[phageData["Virus_ID"]==target_T]["T_nearest_errMar"]))
                match_finder.append(j)
                match_finder.append(statistics.mode(max_finder_match))

    p20_test_matches = pd.DataFrame(np.reshape(np.ravel(match_finder), (-1, 2)))
    p20_test_matches.columns = ['Virus_ID','T_pred']

    # add G2T T
    p20_test_result = p20_test_matches.merge(phageData[['Virus_ID','T_nearest']], how='left', on='Virus_ID')
    # check prediction
    p20_test_result["Correct"]=(p20_test_result["T_nearest"]-p20_test_result["T_pred"])
    
    
   # calculate accuracy
    sim_accuracy = p20_test_result[p20_test_result['Correct']==0].shape[0]
    all_correct.append(sim_accuracy/n20)
    no_prediction = p20_test_matches[p20_test_matches['T_pred']==999].shape[0]
    all_missing.append(no_prediction/n20)
    
avg_missing=np.mean(all_missing)
avg_correct=np.mean(all_correct)
```

```python
# now run it for each increase in similarity requirement

import warnings
warnings.filterwarnings("ignore")

# set train and test sizes
n80 = round(n*.80)
n20 = n-n80
# set number of iterations at each percent similarity
n_iter = 100
# create a random shuffle of the phage data
shuffle_index = np.random.permutation(n)

stats_acc_all = np.zeros((10,3))

for m in range(10):

    all_missing = []
    all_correct = []

    # loop 
    for q in range(n_iter):

        # set train and test sizes
        n80 = round(n*.80)
        n20 = n-n80

        # create a random shuffle of the phage data for this iteration
        shuffle_index = np.random.permutation(shuffle_index)

        # split into an 80/20 training set
        p80 = np.asarray(phageData["Virus_ID"][shuffle_index][:n80])
        p20 = np.asarray(phageData["Virus_ID"][shuffle_index][n80:])

        # break the test set off of the data frame to make the training set
        countcount = []
        for i in range(all_matches.shape[0]):
            if(all_matches.iloc[i]["query to"] in p20):
                #print(T_sim.iloc[i]["query to"])
                countcount.append(i)
        current_train_set = copy.copy(all_matches)
        for i in range (all_matches.shape[0]):
            if i in countcount:
                current_train_set = current_train_set.drop(i)
        current_train_set = current_train_set.reset_index(drop=True)

        # run the test set against the training set for a T-number match
        match_finder = []

        # for each phage in the test set
        for j in p20:
            # look at all matches in the training set
            current_frame = current_train_set[current_train_set['query from']==j]
            current_frame = current_frame[current_train_set['percent identity']>=m*10]

            # if there are no matches, append a nonsense value (999)
            if (current_frame.shape[0]==0):
                    match_finder.append(j)
                    match_finder.append(999)
            # if there are matches, determine the T-number
            else:
                # find the maximum percent identity match from the training set
                max_current_matches = current_frame[current_frame["percent identity"]==max(current_frame["percent identity"])]
                #print(max_current_matches)
                # if there is only one structure in the training set with the max percent match, then use that T-number 
                if (max_current_matches.shape[0]==1):
                    match_finder.append(j)
                    target_T = max_current_matches.iloc[0]['query to']
                    match_finder.append(float(phageData[phageData["Virus_ID"]==target_T]["T_nearest_errMar"]))
                # if there are several matches with the same max percent match, use the consensus of all of the matches for the T-number
                else:
                    max_finder_match = []
                    for k in range(max_current_matches.shape[0]):
                        target_T = max_current_matches.iloc[k]['query to']
                        max_finder_match.append(float(phageData[phageData["Virus_ID"]==target_T]["T_nearest_errMar"]))
                    match_finder.append(j)
                    match_finder.append(statistics.mode(max_finder_match))

        p20_test_matches = pd.DataFrame(np.reshape(np.ravel(match_finder), (-1, 2)))
        p20_test_matches.columns = ['Virus_ID','T_pred']

        # add G2T T
        p20_test_result = p20_test_matches.merge(phageData[['Virus_ID','T_nearest']], how='left', on='Virus_ID')
        # check prediction
        p20_test_result["Correct"]=(p20_test_result["T_nearest"]-p20_test_result["T_pred"])

       # calculate accuracy
        no_prediction = p20_test_matches[p20_test_matches['T_pred']==999].shape[0]
        all_missing.append(no_prediction/n20)
        sim_accuracy = p20_test_result[p20_test_result['Correct']==0].shape[0]
        all_correct.append(sim_accuracy/(n20-no_prediction))
        avg_missing=np.mean(all_missing)
        avg_correct=np.mean(all_correct)
        
    stats_acc_all[m][0]=m*10
    stats_acc_all[m][1]=1-avg_missing
    stats_acc_all[m][2]=avg_correct
    print("loop", m)

print(stats_acc_all)
```
